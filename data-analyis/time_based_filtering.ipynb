## Time Based Filtering

```python
import pandas
from datetime import datetime
from pytz import utc
# reade timestamp as date time 
data=pandas.read_csv('reviews.csv', parse_dates=['Timestamp'])

```

```python
date_filterd_data= data[(data['Timestamp']>datetime(2020,7,1, tzinfo=utc)) &(data['Timestamp']<datetime(2020,12,31, tzinfo=utc))]
date_filterd_data
```
